```python
# Import the dependencies.
import pandas as pd
import numpy as np
import requests
import json
import csv
from pprint import pprint
from config import yelp_api_key
```

```python
# Yelp URL
url = 'https://api.yelp.com/v3/businesses/search'
```

```python
# Pull parameters
# url_params = {'categories':'restaurants', 'location1': 'Denver', 'location2': 'Golden','limit': 10}
url_params = {'categories':'restaurants', 'location': '80216' ,'limit': 50}
# url_params = {'categories':'restaurants', 'location': 'Denver' , 'location[zip_code]':'80216', 'offset': 0, 'limit': 50}
# url_params = {'categories':'restaurants', 'location.zip_code': '80216' ,  'offset': 0, 'limit': 50}
```

```python
# Headers
headers = {
         'Authorization': 'Bearer %s' % yelp_api_key,
    }
```

```python
# Pull JSON 
response = requests.request('GET', url, headers=headers, params=url_params)

pprint(response.json())
```

```python
response.json()['businesses'][0]
```

```python
total_businesses = response.json()['businesses']
```

```python
list = []

for business in total_businesses:
    name = business['name']
    phone = business['phone']
    addresses = business['location']['display_address']
    address = '  '.join(addresses)
    rating = business['rating']
    review_count = business['review_count']
    is_closed = business['is_closed']
    services = business['categories']
    alias = ''
    title = ''
    for service in services:
        alias += service['alias'] + ', '
        title += service['title'] + ', '
        
    lat = business['coordinates']['latitude']
    lon = business['coordinates']['longitude']
    if lat is None:
        continue
    if lon is None:
        continue

    yelp_restaurants = {
                    'name': name,
                    'phone': phone,
                    'address' : address,
                    'category_alias': alias,
                    'category_title': title,
                    'rating' : rating,
                    'review_count' : review_count,
                    'is_closed': is_closed,
                    'lat': lat,
                    'lon': lon,
                        }

    list.append(yelp_restaurants)
    pprint(list)
```

```python
restaurant_df = pd.DataFrame.from_dict(list)
restaurant_df
```

```python
restaurant_df.to_csv('../Resources/restaurant_df_raw50.csv')
```
